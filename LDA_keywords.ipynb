```python
# O problema enfrentado é que alguns artigos não tem keywords
def criaArray(path):
    with open(path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    articles = []
    keywords = []
    
    for line in lines:
        line = line.strip()

        if line.startswith('KW  - '):
            keywords.append(line[6:].lower())
        
        elif line.startswith('ER  -'):
            #se existir keywords no artigo
            if keywords:
                articles.append(keywords)
                # Reinicialize a lista de palavras-chave para o próximo artigo
                keywords = []
    
    return articles

abstracts1 = criaArray(file1)
abstracts2 = criaArray(file2)

abstracts = []
abstracts.extend(abstracts1)
abstracts.extend(abstracts2)

print(abstracts)
```
